## Friends measure

```python
import pandas as pd

# Sample DataFrame with "from" and "to" columns
# data = {'from': [1, 2, 3, 4], 'to': [2, 3, 4, 5]}
# df = pd.DataFrame(data)

# Function to calculate the Friends Measure
def friends_measure(node1, node2, df):
    neighbors1 = set(df[df['from'] == node1]['to'])
    neighbors2 = set(df[df['from'] == node2]['to'])
    
    friends_measure = 1 if neighbors1.intersection(neighbors2) else 0
    
    return friends_measure

# Initialize an empty list to store the Friends Measure for each pair
friends_measures = []

for _, row in combined_df.iterrows():
    from_node = row['from']
    to_node = row['to']
    
    # Calculate the Friends Measure using the function
    friends_score = friends_measure(from_node, to_node, train_data)
    
    # Append the result to the list
    friends_measures.append(friends_score)

# Add the Friends Measure values as a new column to the DataFrame
combined_df['Friends_Measure'] = friends_measures
```

## Oppo dirn frnds

```python
import pandas as pd

# Assuming you have a DataFrame df2 with columns 'from' and 'to'

# Initialize an empty list to store the Opposite-Direction Friends Measure for each pair
opposite_direction_friends = []

for _, row in combined_df.iterrows():
    from_node = row['from']
    to_node = row['to']
    
    # Check if there is a reciprocal connection (opposite direction) in your DataFrame
    reciprocal_connection = train_data[(train_data['from'] == to_node) & (train_data['to'] == from_node)].shape[0]
    
    # Convert the result to 1 or 0
    opposite_direction_friends.append(1 if reciprocal_connection > 0 else 0)

# Add the Opposite-Direction Friends Measure as a new column to df2
combined_df['Opposite_Direction_Friends'] = opposite_direction_friends

# Now df2 contains the Opposite-Direction Friends Measure for each pair of nodes
```
