```python
# Assuming you already have a graph, G, created from your data
# G = nx.from_pandas_edgelist(df1, source='from', target='to', create_using=nx.DiGraph())

def friends_measure(node1, node2, graph):
    # Get the neighbors (neighborhoods) of each node
    neighbors1 = set(graph.neighbors(node1))
    neighbors2 = set(graph.neighbors(node2))
    
    # Initialize the Friends Measure
    friends_measure = 0

    # Calculate the Friends Measure based on the delta function
    for x in neighbors1:
        for y in neighbors2:
            if x == y or graph.has_edge(x, y) or graph.has_edge(y, x):
                friends_measure += 1

    return friends_measure

# Example: Calculate Friends Measure between nodes 1 and 2


# Initialize an empty list to store the Friends Measure for each pair
friends_measures = []

for _, row in combined_df_tf.iterrows():
    from_node = row['from']
    to_node = row['to']
    
    # Calculate the Friends Measure using the function
    friends_score = friends_measure(from_node, to_node, G)  # Assuming G is your graph
    
    # Append the result to the list
    friends_measures.append(friends_score)

# Add the Friends Measure values as a new column to df2
combined_df_tf['Friends_Measure'] = friends_measures

# Now df2 contains the Friends Measure for each pair of nodes
```

```python
import networkx as nx
import community  # You may need to install python-louvain package

# Assuming you have a graph G
# Use the Louvain method to detect communities
partition = community.best_partition(G)

# Create a dictionary mapping nodes to their communities
node_to_community = {node: community_id for node, community_id in partition.items()}

# Assuming you have a DataFrame df2 with columns 'from' and 'to'
same_community_feature = []

for _, row in combined_df_tf.iterrows():
    from_node = row['from']
    to_node = row['to']
    
    # Check if both nodes belong to the same community
    if node_to_community.get(from_node) == node_to_community.get(to_node):
        same_community_feature.append(1)
    else:
        same_community_feature.append(0)

# Add the Same-Community feature as a new column to df2
combined_df_tf['Same_Community'] = same_community_feature

# Now df2 contains the Same-Community feature for each pair of nodes
```
